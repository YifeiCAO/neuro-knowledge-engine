```python
import os, re
import pandas as pd
import numpy as np
from collections import OrderedDict
import xml.etree.ElementTree as ET
from sklearn.feature_extraction.text import CountVectorizer
```

# Load the data

### Link PMIDs to years of publication

```python
years = range(1997, 2019)
```

```python
pmid2year = {}
record_dir = "../../../pubmed/query_180731/records"
missing = []

for file in os.listdir(record_dir):
    pmid = int(file.replace(".xml", ""))
    tree = ET.parse("{}/{}".format(record_dir, file))
    root = tree.getroot()
    try:
        year = root.find("MedlineCitation").find("Article").find("Journal").find("JournalIssue").find("PubDate").find("Year").text
        pmid2year[pmid] = int(year)
    except:
        try:
            date = root.find("MedlineCitation").find("Article").find("Journal").find("JournalIssue").find("PubDate").find("MedlineDate").text
            year = re.search(r"\b(19|20)\d{2}\b", date).group(0)
            pmid2year[pmid] = int(year)
        except:
            print("Missing PMID {}".format(pmid))
            missing.append(pmid)
print("{} articles missing year info".format(len(missing)))
```

```python
pmids = list(pmid2year.keys())
pmid_years = [pmid2year[pmid] for pmid in pmids]
```

```python
df = pd.DataFrame({"PMID": pmids, "YEAR": pmid_years})
df.to_csv("data/pubmed/pubmed_years.csv", index=None)
```

### Compute DTM for PubMed corpus

```python
kmeans_lists = pd.read_csv("../lists/lists/lists_k07_oplen.csv", index_col=None)
rdoc_lists = pd.read_csv("../lists/lists/lists_rdoc_domain_opsim.csv", index_col=None)
```

```python
lexicon = sorted(list(set(kmeans_lists["TOKEN"].append(rdoc_lists["TOKEN"]))))
len(lexicon)
```

```python
pmids = [int(file.replace(".txt", "")) for file in os.listdir("../../../nlp/corpus") if not file.startswith(".")]
```

```python
records = [open("{}/{}.txt".format("../../../nlp/corpus", pmid), "r").read() for pmid in pmids]
```

```python
vec = CountVectorizer(min_df=1, vocabulary=lexicon)
dtm = vec.fit_transform(records)
```

```python
dtm = pd.DataFrame(dtm.toarray(), index=list(pmids), columns=vec.get_feature_names())
dtm = dtm.loc[:, (dtm != 0).any(axis=0)]
dtm.shape
```

## Normalize word counts by article length

```python
doc_lens = {}
for pmid in dtm.index:
    file = "../../../nlp/corpus/{}.txt".format(pmid)
    words = open(file, "r").read().split()
    doc_lens[pmid] = len(words)
```

```python
dtm_norm = np.empty(dtm.shape)
for i, pmid in enumerate(dtm.index):
    dtm_norm[i,:] = dtm.loc[pmid] / float(doc_lens[pmid])
dtm_norm = pd.DataFrame(dtm_norm, index=dtm.index, columns=dtm.columns)
```

## Compute mean word proportion by year

```python
years = range(1997, 2020)
prop = pd.DataFrame(index=dtm.columns, columns=years)
for year in years:
    pmids = df.loc[df["YEAR"] == year, "PMID"]
    prop[year] = dtm_norm.loc[pmids].mean()
```

## Combine words by brain system

### Data-driven domains (*k*=7)

```python
kmeans_doms = [6, 3, 7, 5, 2, 1, 4]
dom2name = {6: "Emotion", 3: "Anticipation", 1: "Meaning", 4: "Language", 
            2: "Manipulation", 5: "Vision", 7: "Cognition"}
kmeans_names = [dom2name[dom] for dom in kmeans_doms]
```

```python
kmeans_prop = pd.DataFrame(index=kmeans_names, columns=years)
for k, name in zip(kmeans_doms, kmeans_names):
    words = kmeans_lists.loc[kmeans_lists["CLUSTER"] == k, "TOKEN"]
    kmeans_prop.loc[name] = prop.loc[words].mean() * 10000
```

```python
kmeans_prop.to_csv("data/pubmed/proportions_kmeans_k07.csv")
```

### RDoC domains

```python
rdoc_doms = OrderedDict.fromkeys(rdoc_lists["DOMAIN"])
rdoc_names = ["Negative Valence", "Positive Valence", "Cognitive Systems",
              "Social Processes", "Arousal & Regulation", "Sensorimotor Systems"]
```

```python
rdoc_prop = pd.DataFrame(index=rdoc_names, columns=years)
for dom, name in zip(rdoc_doms, rdoc_names):
    words = rdoc_lists.loc[rdoc_lists["DOMAIN"] == dom, "TOKEN"]
    rdoc_prop.loc[name] = prop.loc[words].mean() * 10000
```

```python
rdoc_prop.to_csv("data/pubmed/proportions_rdoc_domain.csv")
```

### Before vs. after RDoC

```python
df_before = pd.DataFrame(index=rdoc_names)
df_after = pd.DataFrame(index=rdoc_names)
for i, dom in enumerate(rdoc_names):
    before = rdoc_prop.loc[dom,range(1997,2011)].mean() 
    df_before.loc[dom, "OBSERVED"] = before
    after = rdoc_prop.loc[dom,range(2010,2018)].mean()
    df_after.loc[dom, "OBSERVED"] = before
    print("{:20s} {:6.3f} --> {:6.3f}".format(dom, before, after))
```

```python
n_iter = 1000
boot_before = pd.DataFrame(index=rdoc_names, columns=range(n_iter))
boot_after = pd.DataFrame(index=rdoc_names, columns=range(n_iter))
for n in range(n_iter):
    prop_n = pd.DataFrame(index=dtm.columns, columns=years)
    boot = np.random.choice(range(len(dtm.index)), size=len(dtm.index), replace=True)
    meta_n = df.iloc[boot]
    for year in years:
        pmids = meta_n.loc[meta_n["YEAR"] == year, "PMID"]
        prop_n[year] = dtm_norm.loc[pmids].mean()
    i = 0
    prop_n *= 10000
    for dom, name in zip(rdoc_doms, rdoc_names):
        words = rdoc_lists.loc[rdoc_lists["DOMAIN"] == dom, "TOKEN"]
        boot_before.loc[name, n] = prop_n.loc[words, range(1997,rdoc_tps[i]+1)].mean().mean()
        boot_after.loc[name, n] = prop_n.loc[words, range(rdoc_tps[i], 2018)].mean().mean()
        i += 1
```

### Word-level RDoC proportions

```python
markers = [".", ",", "o", "v", "^", "<", ">", "1", "2", "3", "4", 
           "p", "P", "*", "+", "x", "X", "D", "d", "|", "_", "8",
           "h", "H"]
for i, dom in enumerate(rdoc_doms):
    fig, ax = plt.subplots(figsize=(8, 3))
    words = rdoc_lists.loc[rdoc_lists["DOMAIN"] == dom, "TOKEN"]
    for j, word in enumerate(words):
        plt.scatter(years, prop.loc[word, years] * 10000, label=word, 
                    c=rdoc_palette[i], alpha=0.5, s=40, marker=markers[j])
    
    ax.legend(loc="upper left", prop=prop_sm, edgecolor="white", 
              handletextpad=0.5, borderpad=0.65, bbox_to_anchor=(1, 1.2))
    
    for j, word in enumerate(words):
        plt.plot(years, prop.loc[word, years] * 10000, zorder=0,
                 c=rdoc_palette[i], alpha=0.5, linewidth=2)
    
    maxval = prop.loc[words].max().max() * 10000
    plt.plot([rdoc_tps[i], rdoc_tps[i]], [-1, maxval], 
             c="k", alpha=1, linestyle="dashed", linewidth=2)

    for side in ["right", "top"]:
        ax.spines[side].set_visible(False)

    plt.xticks(range(1997, 2018, 5), fontproperties=prop_lg)
    yint = 1
    if maxval > 10:
        yint = 2
    plt.yticks(range(0, int(maxval)+5, yint), fontproperties=prop_lg)

    plt.ylim([-0.5, maxval+0.5])

    ax.xaxis.set_tick_params(width=2, length=5)
    ax.yaxis.set_tick_params(width=2, length=5)
    
    plt.savefig("figures/pubmed/proportions_rdoc_domain_{}.png".format(dom), 
                dpi=250, bbox_inches="tight")
    plt.show()
```
