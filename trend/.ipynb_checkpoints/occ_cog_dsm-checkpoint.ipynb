```python
import pandas as pd
import numpy as np
from collections import OrderedDict
```

# Load the data

```python
meta = pd.read_csv("../data/metadata_filt_180811.csv", index_col=None)
meta = meta.dropna(subset=["PMID"])
meta["PMID"] = meta["PMID"].astype(int)
meta.head(1)
```

```python
version = 190325
dtm = pd.read_csv("../data/dtm_{}.csv.gz".format(version), 
                  compression="gzip", index_col=0)
```

## Normalize word counts by article length

```python
doc_lens = {}
for pmid in dtm.index:
    file = "../../../nlp/corpus/{}.txt".format(pmid)
    words = open(file, "r").read().split()
    doc_lens[pmid] = len(words)
```

```python
dtm_norm = np.empty(dtm.shape)
for i, pmid in enumerate(dtm.index):
    dtm_norm[i,:] = dtm.loc[pmid] / float(doc_lens[pmid])
dtm_norm = pd.DataFrame(dtm_norm, index=dtm.index, columns=dtm.columns)
```

## Compute mean word proportion by year

```python
years = range(1997, 2018)
prop = pd.DataFrame(index=dtm.columns, columns=years)
for year in years:
    pmids = meta.loc[meta["YEAR"] == year, "PMID"]
    prop[year] = dtm_norm.loc[pmids].mean()
```

## Combine words by brain system

### DSM classes

```python
dsm_lists = pd.read_csv("../lists/lists/lists_dsm_class_opsim.csv", index_col=None)
```

```python
dsm_doms = OrderedDict.fromkeys(dsm_lists["CLASS"])
```

```python
dsm_prop = pd.DataFrame(index=dsm_doms, columns=years)
for dom in dsm_doms:
    words = dsm_lists.loc[dsm_lists["CLASS"] == dom, "TOKEN"]
    dsm_prop.loc[dom] = prop.loc[words].mean() * 10000
```

```python
dsm_prop.to_csv("data/cogneuro/proportions_dsm_class.csv")
```

## Articles in which *RDoC* occurred vs. not

### Word level

```python
dir = "../../../nlp/corpus"
dsm_occ = {pmid: False for pmid in meta["PMID"]}
for pmid in meta["PMID"]:
    infile = "{}/{}.txt".format(dir, pmid)
    text = open(infile, "r").read().split()
    if "dsm" in text:
        dsm_occ[pmid] = True
```

```python
pmids_dsm = {}
pmids_dsm["occ"] = set([pmid for pmid, val in dsm_occ.items() if val])
pmids_dsm["not"] = set(meta["PMID"]).difference(pmids_dsm["occ"])
len(pmids_dsm["occ"]) + len(pmids_dsm["not"])
```

#### Observed values

```python
dsm_words = list(set(dsm_lists["TOKEN"]))
len(dsm_words)
```

```python
part = "occ"
idx = dtm_norm.index.isin(pmids_dsm[part])
np.mean(dtm_norm.values[idx,:], axis=0).shape
```

```python
parts = ["occ", "not"]
dsm_prop = {}
for part in parts:
    dsm_prop[part] = np.mean(dtm_norm.loc[pmids_dsm[part],dsm_words], axis=0) * 10000
    print(part.upper())
    print(dsm_prop[part].sort_values(ascending=False)[:10])
    print("")
```

#### Bootstrap distributions

```python
n_iter = 10000
```

```python
dtm_norms = {part: dtm_norm.loc[pmids_dsm[part], dsm_words] for part in parts}
dsm_prop_boot = {part: np.empty((len(dsm_words), n_iter)) for part in parts}
for n in range(n_iter):
    for part in ["occ", "not"]:
        boot = np.random.choice(range(len(dtm_norms[part])), 
                                size=len(dtm_norms[part]), replace=True)
        dsm_prop_boot[part][:,n] = np.mean(dtm_norms[part].values[boot,:], axis=0) * 10000
```

#### Comparison tests

```python
from statsmodels.stats import multitest
```

```python
pval = []
df = pd.DataFrame()
for i, word in enumerate(dsm_words):   
    dom_dsm = dsm_prop_boot["occ"][i,:]
    dom_not = dsm_prop_boot["not"][i,:]
    if np.mean(dom_dsm) > np.mean(dom_not):
        p = np.sum(np.less(dom_dsm, np.max(dom_not))) / float(n_iter)
    else:
        p = 1 # np.sum(np.less(dom_not, np.max(dom_dsm))) / float(n_iter)
    pval.append(p)
    df.loc[word, "P"] = p
df["P_COR"] = multitest.multipletests(pval, method="fdr_bh")[1]
for word in dsm_words:
    for star, alpha in zip(["*", "**", "***"], [0.05, 0.01, 0.001]):
        if df.loc[word, "P_COR"] < alpha:
            df.loc[word, "STARS"] = star
df = df.loc[dsm_words, ["P", "P_COR", "STARS"]]
```

#### Distribution plots

```python
prop_sm = font_manager.FontProperties(fname=arial, size=18)
prop_md = font_manager.FontProperties(fname=arial, size=20)
prop_lg = font_manager.FontProperties(fname=arial, size=24)
rcParams["axes.linewidth"] = 1.5
```

```python
ymax = [20, 10, 10, 40, 25, 10, 15, 10]
for d, dom in enumerate(dsm_doms):  
    
    # Extract and sort words
    dom_words = set(dsm_lists.loc[dsm_lists["CLASS"] == dom, "TOKEN"])
    dom_words = dom_words.intersection(dsm_prop["occ"].index)
    dom_words = list(dsm_prop["occ"].loc[dom_words].sort_values(ascending=False).index)[:10]
    
    # Set up figure
    fig = plt.figure(figsize=(3.35, 3))
    ax = fig.add_axes([0,0,1,1])
    
    for i, word in enumerate(dom_words):
        idx = dsm_words.index(word)
        
        # Documents in which dsm did NOT occur
        data_not = sorted(dsm_prop_boot["not"][idx,:])
        obs = dsm_prop["not"].loc[word]
        plt.plot([i-0.63, i+0.14], [obs, obs], 
                 c="black", alpha=0.7, lw=2)
        v = ax.violinplot(data_not, positions=[i-0.25], 
                          showmeans=False, showmedians=False, widths=0.85)
        for pc in v["bodies"]:
            pc.set_facecolor("black")
            pc.set_edgecolor("black")
            pc.set_linewidth(1.25)
            pc.set_alpha(0.5)
        for line in ["cmaxes", "cmins", "cbars"]:
            v[line].set_edgecolor("none")

        # Documents in which dsm occurred
        data_occ = sorted(dsm_prop_boot["occ"][idx,:])
        obs = dsm_prop["occ"].loc[word]
        plt.plot([i-0.63, i+0.14], [obs, obs], 
                 c=dsm_palette[d], alpha=1, lw=2)
        v = ax.violinplot(data_occ, positions=[i-0.25], 
                          showmeans=False, showmedians=False, widths=0.85)
        for pc in v["bodies"]:
            pc.set_facecolor(dsm_palette[d])
            pc.set_edgecolor(dsm_palette[d])
            pc.set_linewidth(1.25)
            pc.set_alpha(0.55)
        for line in ["cmaxes", "cmins", "cbars"]:
            v[line].set_edgecolor("none")

        # Comparison test
        alphas = [0.05, 0.01, 0.001]
        dys = [0.0, float(ymax[d])/25.0, float(ymax[d])/12.5]
        dxs = [0.45, 0.45, 0.36, 0.45, 0.45, 0.45, 0.46, 0.45]
        for alpha, dy in zip(alphas, dys):
            if df.loc[word, "P_COR"] < alpha:
                plt.text(i-dxs[d], max(data_not+data_occ) + dy, 
                         "*", fontproperties=prop_lg)
    
    # Set plot parameters
    plt.xticks([v-0.25 for v in range(len(dom_words))], 
               fontproperties=prop_sm)
    ax.set_xticklabels([w.replace("_", " ") for w in dom_words], 
                       rotation=60, ha="right")
    plt.yticks(range(0, ymax[d]+5, 5), fontproperties=prop_md)
    plt.xlim([-1, len(dom_words)-0.5])
    plt.ylim([0, ymax[d]]) 
    for side in ["right", "top"]:
        ax.spines[side].set_visible(False)
    ax.xaxis.set_tick_params(width=1.5, length=7)
    ax.yaxis.set_tick_params(width=1.5, length=7)

    # Export figure
    plt.savefig("figures/cogneuro/dsm_vs_not/{}_{}iter.png".format(dom, n_iter), 
                dpi=250, bbox_inches="tight")
    plt.show()
```

#### Word-level DSM occurrences

```python
dsm_prop_occ_yr = pd.DataFrame(index=dtm_norm.columns, columns=years)
for year in years:
    pmids = set(meta.loc[meta["YEAR"] == year, "PMID"]).intersection(set(pmids_dsm["occ"]))
    dsm_prop_occ_yr[year] = dtm_norm.loc[pmids].mean() * 10000
```

```python
dsm_prop_not_yr = pd.DataFrame(index=dtm.columns, columns=years)
for year in years:
    pmids = set(meta.loc[meta["YEAR"] == year, "PMID"]).intersection(set(pmids_dsm["not"]))
    dsm_prop_not_yr[year] = dtm_norm.loc[pmids].mean() * 10000
```

```python
ymax = [20, 10, 10, 40, 25, 10, 15, 10]
markers = [".", ",", "o", "v", "^", "<", ">", "1", "2", "3", "4", 
           "p", "P", "*", "+", "x", "X", "D", "d", "|", "_", "8",
           "h", "H", 4, 5]
for i, dom in enumerate(dsm_doms):
    fig, ax = plt.subplots(figsize=(4, 4))
    words = dsm_lists.loc[dsm_lists["CLASS"] == dom, "TOKEN"]
    for j, word in enumerate(words):
        plt.scatter(years, dsm_prop_not_yr.loc[word], 
                    c="black", alpha=0.4, marker=markers[j])
    
    ax.legend(loc="upper left", prop=prop_sm, edgecolor="white", 
              handletextpad=0.5, borderpad=0.65, bbox_to_anchor=(1, 1.2))
    
    for j, word in enumerate(words):
        plt.scatter(years, dsm_prop_occ_yr.loc[word], label=word, 
                    c=dsm_palette[i], alpha=0.75, marker=markers[j])
    
    for j, word in enumerate(words):
        plt.plot(years, dsm_prop_not_yr.loc[word], zorder=0,
                 c="black", alpha=0.4, linewidth=2)
        plt.plot(years, dsm_prop_occ_yr.loc[word], zorder=0,
                 c=dsm_palette[i], alpha=0.75, linewidth=2)
        
    for side in ["right", "top"]:
        ax.spines[side].set_visible(False)

    plt.xlim([1996,2018])
    plt.xticks(range(1997,2022,10), fontproperties=prop_md)
    plt.ylim([0, ymax[i]])
    plt.yticks(range(0, ymax[i]+5, 5), fontproperties=prop_md)

    ax.xaxis.set_tick_params(width=1.5, length=7)
    ax.yaxis.set_tick_params(width=1.5, length=7)
    
    plt.savefig("figures/cogneuro/dsm_vs_not/{}_years.png".format(dom), 
                dpi=250, bbox_inches="tight")
    plt.show()
```
